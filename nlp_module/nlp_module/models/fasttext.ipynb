```python
import torch
import torch.nn as nn

# 定义模型
class FastTextModel(nn.Module):
    def __init__(self, vocab_size, embed_size, num_classes):
        super(FastTextModel, self).__init__()
        # 创建 EmbeddingBag 层
        self.embedding = nn.EmbeddingBag(vocab_size, embed_size, mode='mean', sparse=True)  #取平均
        self.fc = nn.Linear(embed_size, num_classes)

    def forward(self, x):
        # 通过 EmbeddingBag 获取嵌入表示
        x = self.embedding(x)
        # 通过全连接层进行分类
        x = self.fc(x)
        return x

```
